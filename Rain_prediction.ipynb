```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
```

```python
# Load / Check the dataset
train_data = pd.read_csv('../data/train.csv')
train_data.head()
```

```python
train_data['RainToday'].describe()
```

```python
train_data.info()
```

As null datasets did not exist, skip to fill in missing values.

```python
# Check numerical data distribution
list(set(train_data.dtypes.tolist()))
train_num = train_data.select_dtypes(include = ['float64', 'int64'])
train_num.hist(figsize=(16, 20), bins=20, xlabelsize=10, ylabelsize=10);
plt.show()
```

```python
# Convert object type into num / Delete id column

train_data2 = train_data.drop('ID', axis=1)
train_data2 = pd.get_dummies(train_data2 ,drop_first = True, dtype = int)
train_data2.shape
```

The number of columns has increased significantly. To find the root cause, export train_data2 to CSV.

```python
train_data2.to_csv('../data/train_data2.csv', index=False)
```

The root cause is that "Date" converted each one-hot data (the number is 2615). Change the type of "Date" from 'object' to 'datetime64' before get_dummies. Also, split 'Date' into 'year'/'month'/'day' as Datetime64 type is not available in DecisionTreeClassifier.

```python
train_data3 = train_data.drop('ID', axis=1)

train_data3['Date'] = pd.to_datetime(train_data3['Date'], format="%y-%m-%d")
train_data3['year'] = train_data3['Date'].dt.year
train_data3['month'] = train_data3['Date'].dt.month
train_data3['day'] = train_data3['Date'].dt.day

rain_today = train_data3['RainToday']
train_data3 = pd.get_dummies(train_data3.drop(columns=['RainToday', 'Date']), drop_first=True, dtype=int)
train_data3['RainToday'] = rain_today

train_data3.to_csv('../data/train_data3.csv', index=False)
train_data3.shape
```

```python
# Set feature values / target variable
X = train_data3.drop('RainToday', axis=1)
y = train_data3['RainToday']

X_train,X_valid,y_train,y_valid = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_valid.head()
```

```python
# Training by DecisionTreeClassifier
# Search for the highest result

from sklearn.tree import DecisionTreeClassifier

for i in range(10):
    model_dtree = DecisionTreeClassifier(max_depth = i+1, random_state = 0, class_weight ='balanced')
    model_dtree.fit(X_train, y_train)
    result = model_dtree.score(X = X_valid,y = y_valid)
    print(f"max_depth: {i+1} result: {result}")
```

'max_depth' from 6 to 8 is a higher value. Tentatively set max_depth is 6 in the model_dtree.

```python
model_dtree = DecisionTreeClassifier(max_depth = 6, random_state = 0, class_weight ='balanced')
model_dtree.fit(X_train, y_train)
model_dtree.score(X = X_valid,y = y_valid)

# Check importance values(top10)
importances = pd.Series(model_dtree.feature_importances_, index=X.columns)
importances.sort_values(ascending=False).head(10)
```

```python
# Search for the best importance values

y2 = train_data3['RainToday']
max_result = 0
max_length = 0

for i in range(len(importances)):
    Xcol2 = importances.sort_values(ascending=False).head(i+1).index.tolist()
    X2 = train_data3[Xcol2]
    
    X_train2,X_valid2,y_train2,y_valid2 = train_test_split(X2, y2, test_size = 0.2, random_state = 0)
    model_dtree2 = DecisionTreeClassifier(max_depth = 6, random_state = 0, class_weight ='balanced')
    model_dtree2.fit(X_train2, y_train2)
    result2 = model_dtree2.score(X = X_valid2, y = y_valid2)

    if result2 > max_result:
        max_length = i+1
        max_result = result2

print(f"max_length: {max_length} max_result: {max_result}")
```

Importance values(top16) is best. Try other methods.

Analyze what type of data was incorrect.

```python
y_pred = model_gboost_final.predict(X_valid2)

results = pd.DataFrame({
    'actual': y_valid2,
    'predicted': y_pred
})

results['correct'] = results['actual'] == results['predicted']
```

```python
# Reset index for joining
results = results.reset_index(drop=True)
X_valid2_reset = X_valid2.reset_index(drop=True)

errors = pd.concat([X_valid2_reset, results], axis=1)

# Filter by the category of confusion matrix
false_positives = errors[(errors['actual'] == 'No') & (errors['predicted'] == 'Yes')]
true_negatives = errors[(errors['actual'] == 'No') & (errors['predicted'] == 'No')]
false_negatives = errors[(errors['actual'] == 'Yes') & (errors['predicted'] == 'No')]
true_positives = errors[(errors['actual'] == 'Yes') & (errors['predicted'] == 'Yes')]

false_positives.to_csv('../reports/false_positives.csv', index=False)
true_negatives.to_csv('../reports/true_negatives.csv', index=False)
false_negatives.to_csv('../reports/false_negatives.csv', index=False)
true_positives.to_csv('../reports/true_positives.csv', index=False)
```

```python
# Compare false_positives vs true_negatives (pred yes/no --> actual no)
for i in Xcol2 :
    plt.figure(figsize=(3, 3))
    plt.hist(false_positives[i], bins=10, alpha=0.5, label='False Positive')
    plt.hist(true_negatives[i], bins=10, alpha=0.5, label='True Negative')
    plt.legend()
    plt.title(f'{i} : FP vs TN')
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```
